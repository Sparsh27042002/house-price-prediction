```python
def convertRange(value):
    try:
        # Check if the value contains a range (e.g., "1200-1500")
        if '-' in value:
            values = value.split('-')
            # Convert both ends of the range to float and calculate the average
            return (float(values[0]) + float(values[1])) / 2
        # If the value is a single number, convert it to float
        return float(value)
    except (ValueError, TypeError):
        # Handle invalid or missing values
        return None
```

```python
def remove_outliers_sqft(df):
    df_output = pd.DataFrame()  # Initialize an empty DataFrame
    for key, subdf in df.groupby('location'):  # Group by 'location'
        m = np.mean(subdf.price_per_sqft)  # Mean of price_per_sqft
        st = np.std(subdf.price_per_sqft)  # Standard deviation of price_per_sqft

        # Filter the sub-dataframe for outliers
        gen_df = subdf[(subdf.price_per_sqft > (m - st)) & (subdf.price_per_sqft <= (m + st))]
        
        # Append filtered data to the output DataFrame
        df_output = pd.concat([df_output, gen_df], ignore_index=True)

    return df_output

# Apply the function
data = remove_outliers_sqft(data)

# Check the result
data.describe()
```

```python
def bhk_outlier_remover(df):
    exclude_indices = np.array([], dtype=int)  # Initialize an empty integer array for indices
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        # Group by 'bhk' within each location
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }

        # Check for outliers
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk - 1)  # Get stats of the previous BHK
            if stats and stats['count'] > 5:  # Ensure valid stats and sufficient data points
                exclude_indices = np.append(
                    exclude_indices,
                    bhk_df[bhk_df.price_per_sqft < (stats['mean'] - stats['std'])].index.values
                )

    # Drop the identified outliers
    return df.drop(exclude_indices, axis='index')
```

```python
data=bhk_outlier_remover(data)
data.sample(10)
```

```python
x=data.drop(columns=['price'])
y=data['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
```

# Applying Linear Regression

```python
column_trans = make_column_transformer(
    (OneHotEncoder(sparse_output=False), ['location']),  # Transformation for 'location'
    remainder='passthrough'                      # Keep other columns as is
)
```

```python
scaler=StandardScaler()
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,scaler,lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred_lr=pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred_lr)
```

# Applying Lasso

```python
lasso=Lasso()
```

```python
pipe=make_pipeline(column_trans,scaler, lasso)
```

```python
 pipe.fit(x_train,y_train)
```

```python
y_pred_lasso=pipe.predict(x_test)
r2_score(y_test,y_pred_lasso)
```

# Applying Ridge

```python
ridge=Ridge()
```

```python
pipe=make_pipeline(column_trans,scaler,ridge)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred_ridge=pipe.predict(x_test)
r2_score(y_test,y_pred_ridge)
```
